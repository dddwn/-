```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import time

# WebDriver 설정
driver = webdriver.Chrome()

# URL로 이동
url = "https://www.kg-mobility.com/kr/service/network/se_pl/index.html"
driver.get(url)

# 데이터 저장할 리스트 초기화
data = []

try:
    # 처음에 맨 아래로 스크롤하여 페이지 전체 로딩을 유도
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)  # 페이지 로딩 대기

    # `iframe`으로 전환
    WebDriverWait(driver, 10).until(EC.frame_to_be_available_and_switch_to_it((By.ID, "service_center")))

    # 1번부터 27번까지의 페이지 번호를 순차적으로 클릭
    for page_num in range(1, 28):
        try:
            # 페이지 번호 버튼 클릭
            page_button = driver.find_element(By.XPATH, f"//div[@id='pagination']/a[contains(@href, 'current_page={page_num}')]")
            driver.execute_script("arguments[0].click();", page_button)
            time.sleep(3)  # 페이지 로딩 대기

            # 데이터 수집
            rows = driver.find_elements(By.XPATH, "//table/tbody/tr")
            for row in rows:
                업체명 = row.find_element(By.XPATH, ".//td[2]").text.strip() or None
                주소 = row.find_element(By.XPATH, ".//td[3]").text.strip() or None
                전화번호 = row.find_element(By.XPATH, ".//td[4]").text.strip() or None
                종류 = "서비스프라자"
                data.append([업체명, 주소, 전화번호, 종류])

        except Exception as page_error:
            print(f"{page_num}번 페이지에서 오류 발생: {page_error}")

except Exception as e:
    print(f"데이터 수집 중 오류 발생: {e}")

# DataFrame으로 변환 및 출력
df2 = pd.DataFrame(data, columns=["업체명", "주소", "전화번호", "종류"])
print(df2)
```

```python
df_combined = pd.concat([df1, df2], axis=0, ignore_index=True)
df_combined
```

```python
df = df_combined
```

```python
df['주소'] = df['주소'].str.replace('특별시|광역시|특별자치도|특별자치시', '', regex=True)
df['주소'] = df['주소'].str.replace('경기도', '경기', regex=False)
df['주소'] = df['주소'].str.replace('강원도', '강원', regex=False)
df['주소'] = df['주소'].str.replace('충청북도', '충북', regex=False)
df['주소'] = df['주소'].str.replace('충청남도', '충남', regex=False)
df['주소'] = df['주소'].str.replace('전라북도', '전북', regex=False)
df['주소'] = df['주소'].str.replace('전라남도', '전남', regex=False)
df['주소'] = df['주소'].str.replace('경상북도', '경북', regex=False)
df['주소'] = df['주소'].str.replace('경상남도', '경남', regex=False)

# '시도' 열에 주소 첫 두 글자 추가
df['시도'] = df['주소'].str[:2]

# '시군구' 열에 첫 번째와 두 번째 띄어쓰기 사이의 텍스트 추출
df['시군구'] = df['주소'].str.extract(r'^\S+\s(\S+)')[0]

df['시도+시군구'] = df['시도'] + df['시군구']
```

```python
import pandas as pd
import numpy as np


# 새로운 열 추가 및 조건에 따라 값 입력
df['보증수리'] = np.where(df['종류'] == '서비스센터', 'o', np.where(df['종류'] == '서비스프라자', 'o', None))
df['일반수리'] = np.where(df['종류'] == '서비스센터', 'o', np.where(df['종류'] == '서비스프라자', 'o', None))
df['사고수리(판금/도색)'] = np.where(df['종류'] == '서비스센터', 'o', None)
df['부품판매'] = np.where(df['종류'] == '서비스센터', 'o', None)
df['용품판매'] = np.where(df['종류'] == '서비스센터', '문의 필요', '문의 필요')

# 결과 확인
print(df)
```
